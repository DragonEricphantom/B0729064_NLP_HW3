```python
import pandas as pd                                             
from sklearn.feature_extraction.text import TfidfVectorizer
wiki = pd.read_csv('movies.csv')

for i in wiki['Label']:
  print(i)
vectorizer = TfidfVectorizer(smooth_idf=True,max_features=20)
tfidf = vectorizer.fit_transform(wiki['Label'])
df_tfidf = pd.DataFrame(tfidf.toarray(),columns=vectorizer.get_feature_names(), index=[wiki['中文名稱']])
print('tfidf = ')
print(tfidf)
df_tfidf

```
